## Visualisation

```python
# matrice de correlation
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

data = Data.drop(columns=["EmployeeCount", "Over18", "StandardHours", "EmployeeNumber"])
cat_columns = ["BusinessTravel", "Department", "EducationField", "Gender", "JobRole", "MaritalStatus" ,"OverTime", "Education", "EnvironmentSatisfaction", "JobInvolvement", "JobLevel", "JobSatisfaction","PerformanceRating", "RelationshipSatisfaction", "StockOptionLevel", "WorkLifeBalance"]
num_columns = data.drop(columns=cat_columns).columns
data["Attrition"] = le.fit_transform(data["Attrition"])

correlation_matrix = data[num_columns].corr()
plt.figure(figsize=(16, 14))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

```python
# création de countplots pour les variables catégorielles par rapport à 'Attrition'

for col in cat_columns:
    plt.figure(figsize=(4,2))
    sns.countplot(data=data, x=col)
    plt.xticks(rotation=45)
    plt.title(f"Distribution de la variable {col}")
```

```python
data = data.drop(columns=[
    "DailyRate",
    "DistanceFromHome",
    "HourlyRate",
    "MonthlyRate",
    "NumCompaniesWorked",
    "YearsSinceLastPromotion",
    "TrainingTimesLastYear",
    "YearsSinceLastPromotion"
    ])

data.columns
```

## Modelisation

```python
from pipeline import split_data, Model_pipeline, metric_model, roc_curve_model

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
# pipeline de logistic regresion
x_train, x_test, y_train, y_test = split_data(data, target="Attrition")
```

```python
# filtre les colonnes
num_columns = [col for col in num_columns if col in x_train.columns]
cat_columns = [col for col in cat_columns if col in x_train.columns]
```

```python
# random de logistic regresion
pipeline_LR = Model_pipeline(num_columns, cat_columns, LogisticRegression())
```

```python
# GridSearch pour le LogisticRegression
param_grid_LR = {
    "model__l1_ratio": [0],
    "model__C": [0.01, 0.1, 1, 10],
    "feature_selection__k": [3, 5, "all"]
    }

grid = GridSearchCV(
    estimator=pipeline_LR,
    param_grid=param_grid_LR)

grid.fit(x_train, y_train)

best_LR = grid.best_estimator_
```

```python
#  metrics de logistic regresion
y_pred_LR = best_LR.predict(x_test)

accuracy, precision, recall, f1 = metric_model(y_test, y_pred_LR)

print("accuracy", accuracy)
print("precision", precision)
print("recall", recall)
print("f1", f1)
```

```python
# matrice de confusion de logistic regresion

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred_LR)

plt.figure(figsize=(6,6))
sns.heatmap(cm, annot=True, cmap="Greens")
plt.title("Matrice de confusion")
plt.ylabel("True positif")
plt.xlabel("Predicted label")
plt.show
```

```python
# pipeline de random forest
pipeline_RF = Model_pipeline(num_columns, cat_columns, RandomForestClassifier())
```

```python
# GridSearch pour le RandomForestClassifier
param_grid_RF = {
    "model__n_estimators": [10, 50, 100, 250, 500],
    "model__max_features": ["sqrt", "log2"],
    "feature_selection__k": [3, 5, "all"]
    }

grid = GridSearchCV(
    estimator=pipeline_RF,
    param_grid=param_grid_RF)

grid.fit(x_train, y_train)

best_RF = grid.best_estimator_
```

```python
#  metrics de random forest
y_pred_RF = best_RF.predict(x_test)

accuracy, precision, recall, f1 = metric_model(y_test, y_pred_RF)

print("accuracy", accuracy)
print("precision", precision)
print("recall", recall)
print("f1", f1)
```

```python
# matrice de confusion de random forest

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred_RF)

plt.figure(figsize=(6,6))
sns.heatmap(cm, annot=True, cmap="Greens")
plt.title("Matrice de confusion")
plt.ylabel("True positif")
plt.xlabel("Predicted label")
plt.show
```

```python
# ROC Logistic Regression
fpr_lr, tpr_lr, auc_lr = roc_curve_model(best_LR, x_test, y_test)

# ROC Random Forest
fpr_rf, tpr_rf, auc_rf = roc_curve_model(best_RF, x_test, y_test)

# Plot compare
plt.figure(figsize=(8, 6))

plt.plot(fpr_lr, tpr_lr, label=f"Logistic Regression (AUC = {auc_lr:.2f})")

plt.plot(fpr_rf, tpr_rf, label=f"Random Forest (AUC = {auc_rf:.2f})")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend()  # pour specifie
plt.show()
```
